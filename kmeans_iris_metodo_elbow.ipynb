### Gráfico do decacimento de erro RMS por clusters
Dados gerados via weka da base iris utilizando até 10 clusters no método kmeans.

No goovy console foi utilizado o script abaixo para a geração dos dados:

```
data = new weka.core.converters.ConverterUtils.DataSource("C:\\iris.arff").getDataSet()
for (int i = 1; i <= 50; i++) {
  kMeans = new weka.clusterers.SimpleKMeans()
  kMeans.setNumClusters(i)
  kMeans.buildClusterer(data)
  System.out.println(i + "\t" + kMeans.getSquaredError())
}
```
fonte do script: https://weka.8497.n7.nabble.com/Identifying-the-elbow-td42951.html

Devido a pouca diferença na variação do erro RMS a partir do cluster 11, apenas os 10 primeiros clusters foram utilizados no gráfico de decaimento.

````python
import matplotlib.pyplot as plt
````

````python
dados = [141.16611042137328, 62.127790750538175, 7.801559361268048, 6.597925743648829, 6.277659330769319, 6.1159421000391125, 
         5.217629646927634, 4.859535068386244, 4.678874159874298, 4.587500225526149]
````

````python
plt.plot(range(1,11),dados, 'go')
plt.plot(range(1,11),dados)
plt.xlim(1, 10)
plt.title('Decaimento do Erro RMS')
plt.xlabel('Clusters')
plt.ylabel('RMS')
plt.show()
````
